##### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target

fig = plt.figure(1, figsize = (6, 6))
ax = Axes3D(fig, elev = 30, azim = -60)

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=Y,
          cmap = plt.cm.rainbow_r)


ax.set_title("PCA directions", fontsize = 20)
ax.set_xlabel("Component 1", fontsize = 15)
ax.set_ylabel("Component 2", fontsize = 15)
ax.set_zlabel("Component 3", fontsize = 15)

plt.show()
```
